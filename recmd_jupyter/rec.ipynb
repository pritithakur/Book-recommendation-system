```python
#Popularity based recommender
```

```python
# merge dataset of books and ratings on the basis of common isbn
name= ratings.merge(books,on='ISBN')
```

```python
#count the ratings of book and rename it
Rating_Num= name.groupby('Book-Title').count()['Book-Rating'].reset_index()
Rating_Num.rename(columns={'Book-Rating':'Rating_Num'},inplace=True)
Rating_Num
```

```python
#average the ratings of books and rename it as avg_rating
Rating_Avg= name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
Rating_Avg.rename(columns={'Book-Rating':'Ratings_Avg'},inplace=True)
Rating_Avg
```

```python
popular = Rating_Num.merge(Rating_Avg,on='Book-Title')
```

```python
popular
```

```python
# Filter num rating with ratings greater than 100
popular=popular[popular['Rating_Num']>=100].sort_values('Ratings_Avg',ascending=False).head(50)
```

```python
#created dataset with ratings greater than 100 and duplicates are dropped
popular = popular.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Rating_Num','Ratings_Avg']]
```

# Collaborative filtering

```python
#merged table of rating with books
name
```

```python
#group user who rated atleast 200 books
x = name.groupby('User-ID').count()['Book-Rating'] > 200
users=x[x].index
```

```python
filter_rating = name[name['User-ID'].isin(users)]
filter_rating
```

```python
#famous books who have rating greater than 50
y=filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
Famous_Book=y[y].index
```

```python
final_rating= filter_rating[filter_rating['Book-Title'].isin(Famous_Book)]
```

```python
st= final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
st.fillna(0,inplace=True)
```

```python
st
#rows as books and columns as users
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#eucledian distance of 706 books with each other
similarity_scores= cosine_similarity(st)
```

```python
similarity_scores
```

```python
def recommend(book_name):
        index = np.where(st.index==book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
        data = []
        for i in similar_items:
            item = []
            temp = books[books['Book-Title'] == st.index[i[0]]]
            item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
            data.append(item)
        print (data)
book_name = input()
if book_name == st.index.any():
    recommend(book_name)
else:
    print('Book not found')
```
